## –ö–∞–∫ —á–µ–ª–æ–≤–µ–∫ —Ä–µ—à–∞–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–µ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –∑–∞ 5-10 —Å–µ–∫—É–Ω–¥?

**–¢–µ–æ—Ä–µ–º–∞ –í–∏–µ—Ç–∞** ‚Äî —ç—Ç–æ –º–æ—â–Ω—ã–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç –¥–ª—è –±—ã—Å—Ç—Ä–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π –º–µ—Ç–æ–¥–æ–º –ø–æ–¥–±–æ—Ä–∞.

–î–ª—è —É—Ä–∞–≤–Ω–µ–Ω–∏—è $x^2 + px + q = 0$:
- **–°—É–º–º–∞ –∫–æ—Ä–Ω–µ–π**: $r_1 + r_2 = -p$
- **–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –∫–æ—Ä–Ω–µ–π**: $r_1 \times r_2 = q$

---

### üìö –ß—Ç–æ –≤—ã —É–∑–Ω–∞–µ—Ç–µ:
1. –ö–∞–∫ —á–µ–ª–æ–≤–µ—á–µ—Å–∫–∏–π –º–æ–∑–≥ –ø–æ–¥–±–∏—Ä–∞–µ—Ç –∫–æ—Ä–Ω–∏
2. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø—Ä–æ—Ü–µ—Å—Å–∞ –º—ã—à–ª–µ–Ω–∏—è
3. –ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–µ —É–ø—Ä–∞–∂–Ω–µ–Ω–∏—è —Å –≥—Ä–∞—Ñ–∏–∫–∞–º–∏
4. –ê–Ω–∏–º–∞—Ü–∏—è –ø—Ä–æ—Ü–µ—Å—Å–∞ –ø–æ–¥–±–æ—Ä–∞

```python
# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫ (—Ç–æ–ª—å–∫–æ –¥–ª—è –ø–µ—Ä–≤–æ–≥–æ –∑–∞–ø—É—Å–∫–∞)
!pip install -q matplotlib numpy ipywidgets plotly
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, IntSlider, widgets
import plotly.graph_objects as go
from IPython.display import display, HTML, clear_output
import time

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Å—Ç–∏–ª—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline
```

## üß† –ö–∞–∫ –¥—É–º–∞–µ—Ç —á–µ–ª–æ–≤–µ–∫ –ø—Ä–∏ —Ä–µ—à–µ–Ω–∏–∏?

–î–∞–≤–∞–π—Ç–µ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã–π –ø—Ä–∏–º–µ—Ä: **$x^2 - 5x - 6 = 0$**

### –•–æ–¥ –º—ã—Å–ª–µ–π:
1. **–û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ü–µ–ª–∏**: –Ω—É–∂–Ω—ã –¥–≤–∞ —á–∏—Å–ª–∞, —Å—É–º–º–∞ –∫–æ—Ç–æ—Ä—ã—Ö **5**, –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ **-6**
2. **–ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –∑–Ω–∞–∫–∏**: –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ ‚Üí –∫–æ—Ä–Ω–∏ —Ä–∞–∑–Ω—ã—Ö –∑–Ω–∞–∫–æ–≤ (–æ–¥–∏–Ω +, –¥—Ä—É–≥–æ–π -)
3. **–ü–µ—Ä–µ–±–∏—Ä–∞–µ–º –¥–µ–ª–∏—Ç–µ–ª–∏** —á–∏—Å–ª–∞ 6: ¬±1, ¬±2, ¬±3, ¬±6
4. **–ü—Ä–æ–≤–µ—Ä—è–µ–º –ø–∞—Ä—ã**:
   - 6 –∏ -1? –°—É–º–º–∞: 6 + (-1) = 5 ‚úÖ –ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ: 6 √ó (-1) = -6 ‚úÖ
   - **–ù–∞—à–ª–∏!** –ö–æ—Ä–Ω–∏: 6 –∏ -1

```python
def visualize_human_thinking(p, q, step_by_step=True):
    """
    –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø—Ä–æ—Ü–µ—Å—Å–∞ –º—ã—à–ª–µ–Ω–∏—è —á–µ–ª–æ–≤–µ–∫–∞ –ø—Ä–∏ –ø–æ–∏—Å–∫–µ –∫–æ—Ä–Ω–µ–π
    –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∫–∞–∂–¥—ã–π —à–∞–≥ —Ä–∞—Å—Å—É–∂–¥–µ–Ω–∏–π
    """
    print("="*60)
    print(f"üéØ –†–µ—à–∞–µ–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ: x¬≤ + ({p})x + ({q}) = 0")
    print("="*60)
    
    target_sum = -p
    target_prod = q
    
    print(f"\nüìå –®–ê–ì 1: –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ü–µ–ª–∏")
    print(f"   –ò—â–µ–º –¥–≤–∞ —á–∏—Å–ª–∞ r‚ÇÅ –∏ r‚ÇÇ:")
    print(f"   ‚Ä¢ r‚ÇÅ + r‚ÇÇ = {target_sum}")
    print(f"   ‚Ä¢ r‚ÇÅ √ó r‚ÇÇ = {target_prod}")
    
    if step_by_step:
        input("\n‚è∏Ô∏è  –ù–∞–∂–º–∏—Ç–µ Enter –¥–ª—è –ø—Ä–æ–¥–æ–ª–∂–µ–Ω–∏—è...")
    
    print(f"\nüîç –®–ê–ì 2: –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –∑–Ω–∞–∫–∏")
    if target_prod > 0:
        print(f"   –ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ {target_prod} > 0 ‚Üí –∫–æ—Ä–Ω–∏ –û–î–ù–û–ì–û –∑–Ω–∞–∫–∞")
        if target_sum > 0:
            print("   –°—É–º–º–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–∞—è ‚Üí –æ–±–∞ –∫–æ—Ä–Ω—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ (+)")
        else:
            print("   –°—É–º–º–∞ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–∞—è ‚Üí –æ–±–∞ –∫–æ—Ä–Ω—è –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ (-)")
    elif target_prod < 0:
        print(f"   –ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ {target_prod} < 0 ‚Üí –∫–æ—Ä–Ω–∏ –†–ê–ó–ù–´–• –∑–Ω–∞–∫–æ–≤ (–æ–¥–∏–Ω +, –¥—Ä—É–≥–æ–π -)")
        if target_sum > 0:
            print("   –°—É–º–º–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–∞—è ‚Üí –±–æ–ª—å—à–∏–π –∫–æ—Ä–µ–Ω—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π")
        else:
            print("   –°—É–º–º–∞ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–∞—è ‚Üí –±–æ–ª—å—à–∏–π –ø–æ –º–æ–¥—É–ª—é –∫–æ—Ä–µ–Ω—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π")
    else:
        print("   –ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ = 0 ‚Üí –æ–¥–∏–Ω –∏–∑ –∫–æ—Ä–Ω–µ–π —Ä–∞–≤–µ–Ω 0")
    
    if step_by_step:
        input("\n‚è∏Ô∏è  –ù–∞–∂–º–∏—Ç–µ Enter –¥–ª—è –ø—Ä–æ–¥–æ–ª–∂–µ–Ω–∏—è...")
    
    # –ù–∞—Ö–æ–¥–∏–º –¥–µ–ª–∏—Ç–µ–ª–∏
    if q == 0:
        divisors = [0, abs(target_sum)]
    else:
        divisors = []
        for i in range(1, abs(q) + 1):
            if q % i == 0:
                divisors.extend([i, -i])
        divisors = sorted(set(divisors))
    
    print(f"\nüìã –®–ê–ì 3: –ù–∞—Ö–æ–¥–∏–º –¥–µ–ª–∏—Ç–µ–ª–∏ —á–∏—Å–ª–∞ {q}")
    print(f"   –î–µ–ª–∏—Ç–µ–ª–∏: {divisors}")
    
    if step_by_step:
        input("\n‚è∏Ô∏è  –ù–∞–∂–º–∏—Ç–µ Enter –¥–ª—è –Ω–∞—á–∞–ª–∞ –ø–µ—Ä–µ–±–æ—Ä–∞...")
    
    print(f"\nüîÑ –®–ê–ì 4: –ü–µ—Ä–µ–±–∏—Ä–∞–µ–º –ø–∞—Ä—ã –¥–µ–ª–∏—Ç–µ–ª–µ–π")
    print("   (–ö–∞–∫ –¥—É–º–∞–µ—Ç —á–µ–ª–æ–≤–µ–∫: –ø—Ä–æ–≤–µ—Ä—è–µ–º –ø–∞—Ä—ã –æ–¥–Ω—É –∑–∞ –¥—Ä—É–≥–æ–π)\n")
    
    attempt = 0
    found_roots = None
    
    for i, r1 in enumerate(divisors):
        if r1 == 0 and q != 0:
            continue
        
        r2 = q / r1 if r1 != 0 else 0
        
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ r2 —Ü–µ–ª–æ–µ –∏ –≤—Ö–æ–¥–∏—Ç –≤ –¥–µ–ª–∏—Ç–µ–ª–∏
        if r2 == int(r2):
            r2 = int(r2)
            attempt += 1
            
            current_sum = r1 + r2
            current_prod = r1 * r2
            
            print(f"   –ü–æ–ø—ã—Ç–∫–∞ {attempt}: r‚ÇÅ = {r1:3}, r‚ÇÇ = {r2:3}")
            print(f"      –°—É–º–º–∞: {r1} + {r2} = {current_sum:3} ", end="")
            
            if current_sum == target_sum:
                print("‚úÖ –ü–æ–¥—Ö–æ–¥–∏—Ç!")
            else:
                print(f"‚ùå –ù—É–∂–Ω–æ {target_sum}")
            
            print(f"      –ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ: {r1} √ó {r2} = {current_prod:3} ", end="")
            
            if current_prod == target_prod:
                print("‚úÖ –ü–æ–¥—Ö–æ–¥–∏—Ç!")
            else:
                print(f"‚ùå –ù—É–∂–Ω–æ {target_prod}")
            
            if current_sum == target_sum and current_prod == target_prod:
                print(f"\n   üéâ –ù–ê–®–õ–ò! –ö–æ—Ä–Ω–∏: {r1} –∏ {r2}")
                found_roots = sorted([r1, r2])
                break
            
            print()
            
            if step_by_step and attempt < len(divisors):
                time.sleep(0.3)  # –ù–µ–±–æ–ª—å—à–∞—è –ø–∞—É–∑–∞ –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏
    
    print("\n" + "="*60)
    if found_roots:
        print(f"‚úÖ –û–¢–í–ï–¢: x‚ÇÅ = {found_roots[0]}, x‚ÇÇ = {found_roots[1]}")
        print(f"   –ü—Ä–æ–≤–µ—Ä–∫–∞: ({found_roots[0]}) + ({found_roots[1]}) = {sum(found_roots)} (–¥–æ–ª–∂–Ω–æ –±—ã—Ç—å {target_sum})")
        print(f"   –ü—Ä–æ–≤–µ—Ä–∫–∞: ({found_roots[0]}) √ó ({found_roots[1]}) = {found_roots[0] * found_roots[1]} (–¥–æ–ª–∂–Ω–æ –±—ã—Ç—å {target_prod})")
    else:
        print("‚ùå –¶–µ–ª—ã—Ö –∫–æ—Ä–Ω–µ–π –Ω–µ –Ω–∞–π–¥–µ–Ω–æ (–≤–æ–∑–º–æ–∂–Ω–æ, –∫–æ—Ä–Ω–∏ –¥—Ä–æ–±–Ω—ã–µ –∏–ª–∏ –∏—Ä—Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω—ã–µ)")
    print("="*60)
    
    return found_roots
```

```python
# –ü—Ä–∏–º–µ—Ä: x¬≤ - 5x - 6 = 0
print("üéì –î–ï–ú–û–ù–°–¢–†–ê–¶–ò–Ø: –ö–∞–∫ —á–µ–ª–æ–≤–µ–∫ —Ä–µ—à–∞–µ—Ç —É—Ä–∞–≤–Ω–µ–Ω–∏–µ\n")
roots = visualize_human_thinking(p=-5, q=-6, step_by_step=False)
```

## üìä –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è: –ì—Ä–∞—Ñ–∏–∫ –ø–∞—Ä–∞–±–æ–ª—ã –∏ –∫–æ—Ä–Ω–∏

–î–∞–≤–∞–π—Ç–µ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ **–≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–π —Å–º—ã—Å–ª** –∫–æ—Ä–Ω–µ–π!

```python
def plot_parabola_with_roots(p, q):
    """
    –†–∏—Å—É–µ—Ç –ø–∞—Ä–∞–±–æ–ª—É y = x¬≤ + px + q –∏ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∫–æ—Ä–Ω–∏
    """
    # –ù–∞—Ö–æ–¥–∏–º –∫–æ—Ä–Ω–∏
    discriminant = p**2 - 4*q
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # –ì—Ä–∞—Ñ–∏–∫ –ø–∞—Ä–∞–±–æ–ª—ã
    x = np.linspace(-10, 10, 500)
    y = x**2 + p*x + q
    
    ax1.plot(x, y, 'b-', linewidth=2, label=f'y = x¬≤ + ({p})x + ({q})')
    ax1.axhline(y=0, color='k', linestyle='--', alpha=0.3)
    ax1.axvline(x=0, color='k', linestyle='--', alpha=0.3)
    ax1.grid(True, alpha=0.3)
    ax1.set_xlabel('x', fontsize=12)
    ax1.set_ylabel('y', fontsize=12)
    ax1.set_title('–ì—Ä–∞—Ñ–∏–∫ –ø–∞—Ä–∞–±–æ–ª—ã', fontsize=14, fontweight='bold')
    
    if discriminant >= 0:
        sqrt_d = np.sqrt(discriminant)
        r1 = (-p + sqrt_d) / 2
        r2 = (-p - sqrt_d) / 2
        
        # –û—Ç–º–µ—á–∞–µ–º –∫–æ—Ä–Ω–∏
        ax1.plot([r1, r2], [0, 0], 'ro', markersize=12, label=f'–ö–æ—Ä–Ω–∏: {r1:.2f}, {r2:.2f}')
        ax1.plot([r1, r1], [0, -1], 'r--', alpha=0.5)
        ax1.plot([r2, r2], [0, -1], 'r--', alpha=0.5)
        
        # –í–µ—Ä—à–∏–Ω–∞ –ø–∞—Ä–∞–±–æ–ª—ã
        vertex_x = -p / 2
        vertex_y = vertex_x**2 + p*vertex_x + q
        ax1.plot(vertex_x, vertex_y, 'go', markersize=10, label=f'–í–µ—Ä—à–∏–Ω–∞: ({vertex_x:.2f}, {vertex_y:.2f})')
        
        ax1.legend(fontsize=10)
        
        # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –Ω–∞ —á–∏—Å–ª–æ–≤–æ–π –ø—Ä—è–º–æ–π
        ax2.axhline(y=0, color='k', linewidth=2)
        ax2.plot([r1, r2], [0, 0], 'ro', markersize=15)
        
        # –î–æ–±–∞–≤–ª—è–µ–º —Å—Ç—Ä–µ–ª–∫–∏ –∏ –ø–æ–¥–ø–∏—Å–∏
        ax2.annotate(f'r‚ÇÅ = {r1:.2f}', xy=(r1, 0), xytext=(r1, 0.5),
                    fontsize=12, ha='center',
                    arrowprops=dict(arrowstyle='->', color='red', lw=2))
        ax2.annotate(f'r‚ÇÇ = {r2:.2f}', xy=(r2, 0), xytext=(r2, -0.5),
                    fontsize=12, ha='center',
                    arrowprops=dict(arrowstyle='->', color='red', lw=2))
        
        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º —Å—É–º–º—É –∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ
        mid_point = (r1 + r2) / 2
        ax2.plot([r1, r2], [0.8, 0.8], 'g-', linewidth=2)
        ax2.text(mid_point, 1.0, f'–°—É–º–º–∞ = {r1+r2:.2f}\n(–¥–æ–ª–∂–Ω–æ –±—ã—Ç—å {-p})',
                ha='center', fontsize=11, bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.7))
        
        ax2.text(mid_point, -1.3, f'–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ = {r1*r2:.2f}\n(–¥–æ–ª–∂–Ω–æ –±—ã—Ç—å {q})',
                ha='center', fontsize=11, bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.7))
        
        ax2.set_xlim(min(r1, r2) - 2, max(r1, r2) + 2)
        ax2.set_ylim(-2, 2)
        ax2.set_title('–ö–æ—Ä–Ω–∏ –Ω–∞ —á–∏—Å–ª–æ–≤–æ–π –ø—Ä—è–º–æ–π', fontsize=14, fontweight='bold')
        ax2.grid(True, alpha=0.3)
        ax2.set_yticks([])
    else:
        ax1.text(0, 0, '–ù–µ—Ç –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã—Ö –∫–æ—Ä–Ω–µ–π\n(–¥–∏—Å–∫—Ä–∏–º–∏–Ω–∞–Ω—Ç < 0)',
                ha='center', fontsize=14, bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))
        ax2.text(0, 0, '–ö–æ—Ä–Ω–∏ –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã–µ',
                ha='center', fontsize=14, bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))
    
    plt.tight_layout()
    plt.show()
```

```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–ª—è x¬≤ - 5x - 6 = 0
plot_parabola_with_roots(p=-5, q=-6)
```

## üéÆ –ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω–æ–µ —É–ø—Ä–∞–∂–Ω–µ–Ω–∏–µ

–¢–µ–ø–µ—Ä—å **–í–´** –º–æ–∂–µ—Ç–µ –∏–∑–º–µ–Ω—è—Ç—å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∏ —Å—Ä–∞–∑—É –≤–∏–¥–µ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç!

```python
def interactive_vieta(p, q):
    """
    –ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω–∞—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å –ø–æ–ª–∑—É–Ω–∫–∞–º–∏
    """
    print(f"\n{'='*60}")
    print(f"üìê –£—Ä–∞–≤–Ω–µ–Ω–∏–µ: x¬≤ + ({p})x + ({q}) = 0")
    print(f"{'='*60}\n")
    
    target_sum = -p
    target_prod = q
    
    print(f"üéØ –¶–µ–ª—å: –Ω–∞–π—Ç–∏ r‚ÇÅ –∏ r‚ÇÇ")
    print(f"   –°—É–º–º–∞: r‚ÇÅ + r‚ÇÇ = {target_sum}")
    print(f"   –ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ: r‚ÇÅ √ó r‚ÇÇ = {target_prod}\n")
    
    # –†–µ—à–∞–µ–º —á–µ—Ä–µ–∑ –¥–∏—Å–∫—Ä–∏–º–∏–Ω–∞–Ω—Ç –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏
    discriminant = p**2 - 4*q
    
    if discriminant >= 0:
        sqrt_d = np.sqrt(discriminant)
        r1 = (-p + sqrt_d) / 2
        r2 = (-p - sqrt_d) / 2
        print(f"‚úÖ –ö–æ—Ä–Ω–∏: r‚ÇÅ = {r1:.3f}, r‚ÇÇ = {r2:.3f}")
        
        if r1 == int(r1) and r2 == int(r2):
            print("   ‚≠ê –¶–µ–ª—ã–µ –∫–æ—Ä–Ω–∏ - –∏–¥–µ–∞–ª—å–Ω–æ –¥–ª—è –º–µ—Ç–æ–¥–∞ –í–∏–µ—Ç–∞!")
        else:
            print("   ‚ö†Ô∏è  –ö–æ—Ä–Ω–∏ –Ω–µ —Ü–µ–ª—ã–µ - –º–µ—Ç–æ–¥ –í–∏–µ—Ç–∞ —Å–ª–æ–∂–Ω–µ–µ –ø—Ä–∏–º–µ–Ω–∏—Ç—å")
    else:
        print(f"‚ùå –ù–µ—Ç –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã—Ö –∫–æ—Ä–Ω–µ–π (–¥–∏—Å–∫—Ä–∏–º–∏–Ω–∞–Ω—Ç = {discriminant} < 0)")
    
    print("\n" + "="*60)
    
    # –ì—Ä–∞—Ñ–∏–∫
    plot_parabola_with_roots(p, q)

# –°–æ–∑–¥–∞—ë–º –∏–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–µ –ø–æ–ª–∑—É–Ω–∫–∏
interact(interactive_vieta, 
         p=IntSlider(min=-10, max=10, step=1, value=-5, description='p:'),
         q=IntSlider(min=-20, max=20, step=1, value=-6, description='q:'));
```

## üèãÔ∏è –¢—Ä–µ–Ω–∞–∂–µ—Ä: –†–µ—à–∏—Ç–µ —Å–∞–º–∏!

–ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Ä–µ—à–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ —É—Ä–∞–≤–Ω–µ–Ω–∏—è **–º–µ—Ç–æ–¥–æ–º –í–∏–µ—Ç–∞** –∑–∞ 10-15 —Å–µ–∫—É–Ω–¥ –∫–∞–∂–¥–æ–µ!

```python
import random

def generate_training_problems(n=5):
    """
    –ì–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–µ –∑–∞–¥–∞—á–∏ —Å —Ü–µ–ª—ã–º–∏ –∫–æ—Ä–Ω—è–º–∏
    """
    print("üèãÔ∏è –¢–†–ï–ù–ê–ñ–ï–†: –†–µ—à–∏—Ç–µ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –º–µ—Ç–æ–¥–æ–º –í–∏–µ—Ç–∞\n")
    print("="*60)
    
    problems = []
    
    for i in range(n):
        # –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Ü–µ–ª—ã–µ –∫–æ—Ä–Ω–∏
        r1 = random.randint(-8, 8)
        r2 = random.randint(-8, 8)
        
        # –í—ã—á–∏—Å–ª—è–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã
        p = -(r1 + r2)
        q = r1 * r2
        
        problems.append((p, q, r1, r2))
        
        print(f"\n{i+1}. x¬≤ + ({p})x + ({q}) = 0")
        print(f"   –ü–æ–¥—Å–∫–∞–∑–∫–∞: —Å—É–º–º–∞ –∫–æ—Ä–Ω–µ–π = {-p}, –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ = {q}")
    
    print("\n" + "="*60)
    
    show_answers = input("\n–ü–æ–∫–∞–∑–∞—Ç—å –æ—Ç–≤–µ—Ç—ã? (y/n): ").lower()
    
    if show_answers == 'y':
        print("\nüìù –û–¢–í–ï–¢–´:\n")
        for i, (p, q, r1, r2) in enumerate(problems, 1):
            roots = sorted([r1, r2])
            print(f"{i}. –ö–æ—Ä–Ω–∏: {roots[0]} –∏ {roots[1]}")
            print(f"   –ü—Ä–æ–≤–µ—Ä–∫–∞: {roots[0]} + {roots[1]} = {sum(roots)} (—Å—É–º–º–∞)")
            print(f"            {roots[0]} √ó {roots[1]} = {roots[0]*roots[1]} (–ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ)\n")

generate_training_problems(5)
```

```python
# –í—ã–±–µ—Ä–∏—Ç–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–ª—è –¥–µ—Ç–∞–ª—å–Ω–æ–≥–æ —Ä–∞–∑–±–æ—Ä–∞
print("–ü—Ä–∏–º–µ—Ä—ã —É—Ä–∞–≤–Ω–µ–Ω–∏–π –¥–ª—è —Ä–∞–∑–±–æ—Ä–∞:")
print("1. x¬≤ - 5x - 6 = 0   (–∫–ª–∞—Å—Å–∏—á–µ—Å–∫–∏–π –ø—Ä–∏–º–µ—Ä)")
print("2. x¬≤ - 3x + 2 = 0   (–æ–±–∞ –∫–æ—Ä–Ω—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ)")
print("3. x¬≤ + 2x - 8 = 0   (–∫–æ—Ä–Ω–∏ —Ä–∞–∑–Ω—ã—Ö –∑–Ω–∞–∫–æ–≤)")
print("4. x¬≤ - 7x + 12 = 0  (–ø–æ–±–æ–ª—å—à–µ –¥–µ–ª–∏—Ç–µ–ª–µ–π)")
print("5. –í–≤–µ—Å—Ç–∏ —Å–≤–æ–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã\n")

choice = input("–í–∞—à –≤—ã–±–æ—Ä (1-5): ")

equations = {
    '1': (-5, -6),
    '2': (-3, 2),
    '3': (2, -8),
    '4': (-7, 12)
}

if choice in equations:
    p, q = equations[choice]
elif choice == '5':
    p = int(input("–í–≤–µ–¥–∏—Ç–µ p: "))
    q = int(input("–í–≤–µ–¥–∏—Ç–µ q: "))
else:
    p, q = -5, -6

print("\n" + "="*60)
print("–ü–û–®–ê–ì–û–í–´–ô –†–ê–ó–ë–û–†")
print("="*60 + "\n")

visualize_human_thinking(p, q, step_by_step=True)
```
